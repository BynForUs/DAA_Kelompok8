```python
# --- Jawaban Tambahan ---
"""
Pendekatan/Algoritma:
1.  Pendekatan yang digunakan adalah *Iterasi Linier (Linear Iteration)*.
2.  Kita menginisialisasi penghitung ke nol dan mengulang melalui setiap elemen.
3.  Untuk menentukan genap, kita menggunakan *operator modulo (%)*. Jika bilangan dibagi 2 sisanya 0 (bilangan % 2 == 0), maka bilangan itu genap.

Kasus Khusus (Edge Cases):
1.  *Array Kosong ([]):* Fungsi harus mengembalikan *0*, yang ditangani dengan baik karena loop tidak akan berjalan.
2.  *Angka Nol (0):* Angka nol dianggap genap, dan (0 % 2 == 0) akan mengembalikan True.
"""
def hitung_angka_genap(arr):
    """
    Menghitung berapa banyak angka genap dalam sebuah array/list.
    """
    jumlah_genap = 0
    # Iterasi melalui setiap elemen
    for bilangan in arr:
        # Pengecekan genap menggunakan operator modulo
        if bilangan % 2 == 0:
            jumlah_genap += 1
    return jumlah_genap

# Contoh
input_arr_1 = [1, 2, 4, 7, 10]
print(f"Input: {input_arr_1}, Output: {hitung_angka_genap(input_arr_1)}") # Output: 3
print(f"Input kosong: [], Output: {hitung_angka_genap([])}") # Output: 0
```

```python
# --- Jawaban Tambahan ---
"""
Langkah-langkah Penyelesaian:
1.  *Penanganan Kasus Kosong:* Periksa apakah array kosong. Jika ya, kembalikan indikator khusus (misalnya, None).
2.  *Inisialisasi:* Asumsikan elemen pertama adalah nilai terbesar sementara.
3.  *Perbandingan:* Ulangi melalui sisa array, bandingkan setiap elemen dengan nilai terbesar sementara.
4.  *Pembaruan:* Jika elemen saat ini lebih besar, jadikan elemen itu nilai terbesar yang baru.

Apa yang harus dilakukan jika array kosong?
1.  Fungsi harus mengembalikan indikasi yang jelas bahwa tidak ada nilai terbesar yang valid, seperti *None* (seperti yang dilakukan dalam kode di atas) atau menaikkan pengecualian (raise exception).
"""
def cari_nilai_terbesar(arr):
    """
    Menemukan nilai terbesar dari sebuah array/list.
    """
    # Kasus khusus: Array kosong
    if not arr:
        # Mengembalikan None untuk menunjukkan ketiadaan nilai
        return None

    # Inisialisasi nilai terbesar sementara dengan elemen pertama
    nilai_terbesar = arr[0]

    # Iterasi dimulai dari elemen kedua (indeks 1)
    for bilangan in arr[1:]:
        if bilangan > nilai_terbesar:
            nilai_terbesar = bilangan

    return nilai_terbesar

# Contoh
input_arr_2 = [5, 2, 9, 1]
print(f"Input: {input_arr_2}, Output: {cari_nilai_terbesar(input_arr_2)}") # Output: 9

# Kasus Khusus
input_arr_kosong = []
print(f"Input kosong: {input_arr_kosong}, Output: {cari_nilai_terbesar(input_arr_kosong)}") # Output: None
```

```python
# --- Jawaban Tambahan ---
"""
Logika yang digunakan untuk membalik string:
1.  Menggunakan *Iterasi Mundur (Reverse Iteration)*.
2.  Kita menginisialisasi string kosong.
3.  Looping dimulai dari karakter terakhir string input dan berakhir di karakter pertama.
4.  Pada setiap langkah perulangan, karakter dari string input ditambahkan ke akhir string hasil. Ini secara efektif menyusun ulang karakter dalam urutan terbalik.

Catatan (Alternatif Python yang lebih ringkas):
Metode Python yang lebih idiomatis adalah menggunakan slicing: return s[::-1]
"""
def balik_string(s):
    """
    Membalik urutan karakter dalam sebuah string.
    """
    string_terbalik = ""
    # Looping mundur, dari indeks terakhir (len(s) - 1) hingga -1 (indeks sebelum 0)
    for i in range(len(s) - 1, -1, -1):
        string_terbalik += s[i]
    return string_terbalik

# Contoh
input_str_3 = "halo"
print(f"Input: '{input_str_3}', Output: '{balik_string(input_str_3)}'") # Output: 'olah'
```

```python
# --- Jawaban Tambahan ---
"""
Bagaimana fungsi ini bekerja langkah demi langkah:
1.  *Inisialisasi Akumulator:* Variabel total_jumlah diatur ke *0*. Ini adalah tempat penyimpanan hasil penjumlahan.
2.  *Iterasi:* Fungsi mulai mengulang elemen-elemen dalam array.
3.  *Akumulasi:* Di setiap langkah, elemen array saat ini *ditambahkan* ke nilai total_jumlah yang ada.
    * Misalnya: Mulai (0) + 3 = 3. Kemudian 3 + 1 = 4. Kemudian 4 + 2 = 6.
4.  *Pengembalian Hasil:* Setelah semua elemen selesai diproses, nilai akhir dari akumulator dikembalikan.
"""
def jumlahkan_elemen(arr):
    """
    Mengembalikan jumlah seluruh elemen di dalam array bilangan bulat.
    """
    total_jumlah = 0

    for bilangan in arr:
        # Akumulasi nilai
        total_jumlah += bilangan

    return total_jumlah

# Contoh
input_arr_4 = [3, 1, 2]
print(f"Input: {input_arr_4}, Output: {jumlahkan_elemen(input_arr_4)}") # Output: 6
```

```python
# --- Jawaban Tambahan ---
"""
Cara Menentukan Apakah String Tersebut Palindrome:
1.  Menggunakan pendekatan *Dua Arah (Two-Pointer)* atau *Pencerminan Setengah Iterasi*.
2.  Loop dilakukan hanya sampai *setengah* dari panjang string.
3.  Di setiap langkah, kita membandingkan karakter pada posisi i dari *depan* dengan karakter pada posisi yang dicerminkan dari *belakang* (n - 1 - i).
4.  Jika ada ketidakcocokan yang ditemukan pada perbandingan mana pun, string *bukan* palindrome, dan kita segera mengembalikan False.
5.  Jika loop selesai tanpa ada ketidakcocokan, berarti string adalah palindrome, dan kita mengembalikan True.
"""
def cek_palindrome(s):
    """
    Memeriksa apakah sebuah string merupakan palindrome.
    (Versi ini mengasumsikan input hanya huruf kecil dan tanpa spasi untuk kesederhanaan)
    """
    n = len(s)
    # Kita hanya perlu mengulang sampai setengah string
    for i in range(n // 2): # n // 2 adalah pembagian bulat (floor division)
        # Bandingkan karakter dari depan (i) dengan karakter dari belakang (n - 1 - i)
        if s[i] != s[n - 1 - i]:
            return False # Bukan Palindrome

    return True # Palindrome

# Contoh
print(f"Input: 'level', Output: {cek_palindrome('level')}") # Output: True
print(f"Input: 'hello', Output: {cek_palindrome('hello')}") # Output: False
```

```python
# --- Jawaban Tambahan ---
"""
Jelaskan Metode Pencarian yang Kamu Gunakan:
1.  Metode pencarian yang digunakan adalah *Pencarian Linier (Linear Search)*.
2.  *Cara Kerja:* Pencarian Linier bekerja dengan memeriksa *setiap elemen* dalam array secara berurutan, satu per satu, mulai dari awal.
3.  *Keuntungan:* Sederhana dan efektif untuk array yang *tidak terurut*.
4.  *Efisiensi:* Jika target ditemukan, proses akan berhenti segera. Jika tidak, proses harus berjalan hingga akhir array.
"""
def temukan_elemen(arr, target):
    """
    Menentukan apakah nilai target terdapat dalam array.
    """
    # Iterasi melalui array
    for bilangan in arr:
        # Bandingkan dengan target
        if bilangan == target:
            return True # Ditemukan, segera kembalikan True

    # Jika loop selesai tanpa menemukan target
    return False

# Contoh
array_7 = [4, 7, 1, 9]
target_7_ada = 7
target_7_tidak_ada = 5

print(f"Array: {array_7}, Target: {target_7_ada}, Output: {temukan_elemen(array_7, target_7_ada)}") # Output: True
print(f"Array: {array_7}, Target: {target_7_tidak_ada}, Output: {temukan_elemen(array_7, target_7_tidak_ada)}") # Output: False
```
