```python
class graph:
    def __init__(self, gdict=None):
        if gdict is None:
            gdict = {}
        self.gdict = gdict

    def getVertices(self):
        return list(self.gdict.keys())

    def addVertex(self, vrtx):
        if vrtx not in self.gdict:
            self.gdict[vrtx] = []

    # Fungsi tambahan untuk membuat garis penghubung (edge)
    def addEdge(self, vrtx1, vrtx2):
        if vrtx1 in self.gdict:
            self.gdict[vrtx1].append(vrtx2)
        if vrtx2 in self.gdict:
            self.gdict[vrtx2].append(vrtx1)

    def getEdges(self):
        # Fungsi bantu untuk melihat koneksi
        return self.gdict

# 1. Membuat Graph awal (bagian r-o-m dan l-p-n sesuai gambar)
graph_elements = {
    "r": ["o", "l"],
    "o": ["r", "m", "p"],
    "m": ["o", "n"],
    "l": ["r", "p"],
    "p": ["l", "o", "n"],
    "n": ["m", "p"]
}

g = graph(graph_elements)

print("--- Sebelum ditambah K dan J ---")
print(g.getVertices())

# 2. MENAMBAH SIMPUL K DAN J
g.addVertex("K")
g.addVertex("J")

# 3. MEMBUAT KONEKSI (EDGE) SESUAI GAMBAR
# K terhubung dengan l
g.addEdge("l", "K")

# J terhubung dengan n
g.addEdge("n", "J")

print("\n--- Setelah ditambah K dan J (dan dihubungkan) ---")
print("Simpul:", g.getVertices())
print("\nDetail Koneksi (Dictionary):")
print(g.getEdges())
```
